```python
import pandas as pd
data = pd.read_csv("/content/spambase.data")
print(data.head())
print(data.dtypes)
print(data.describe())
```

```python
X = data.drop(columns = ['1']) # features
Y = data['1'] #target
```

```python
from sklearn.model_selection import train_test_split

# Assume 'data' is your DataFrame and '1' is the target column
X = data.drop(columns=['1'])  # Features
Y = data['1']  # Target

# Split the data: 80% training, 20% testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Verify the sizes of the splits
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

```python
from sklearn.naive_bayes import GaussianNB

# Step 1: Create a GaussianNB classifier object
nb_model = GaussianNB()

# Step 2: Train the model using the training data
nb_model.fit(X_train, Y_train)

# Verify the model has been trained
print("Model training complete.")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

# Assume 'data' is your DataFrame and '1' is the target column
X = data.drop(columns=['1'])  # Features
Y = data['1']  # Target

# Split the data: 80% training, 20% testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create a GaussianNB classifier object
nb_model = GaussianNB()

# Train the model using the training data
nb_model.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = nb_model.predict(X_test)
print("Predictions on the test set:")
print(Y_pred)

# Predict the probabilities of each class
Y_prob = nb_model.predict_proba(X_test)
print("Predicted probabilities on the test set:")
print(Y_prob)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Step 1: Calculate the accuracy score
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Step 2: Draw the confusion matrix
conf_matrix = confusion_matrix(Y_test, Y_pred)

# Step 3: Plot the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# Assume 'data' is your DataFrame and '1' is the target column
X = data.drop(columns=['1'])  # Features
Y = data['1']  # Target

# Split the data: 80% training, 20% testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create a Linear Regression model object
reg_model = LinearRegression()

# Train the model using the training data
reg_model.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = reg_model.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Mean Absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared Score: {r2:.2f}")

# Plot true vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, Y_pred, color='blue', edgecolor='k', alpha=0.7)
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], color='red', linestyle='--')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True Values vs Predictions')
plt.show()
```
