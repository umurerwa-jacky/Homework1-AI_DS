```python
import matplotlib.pyplot as plt
import pandas as pd

# Data dictionary for simplicity
data = {
    "District": ["Nyarugenge", "Gasabo", "Kicukiro", "Nyanza", "Gigasara", "Nyaruguru", "Huye", "Nyamagabe", "Ruhango", 
                 "Muhanga", "Kamonyi", "Karongi", "Rutsiro", "Rubavu", "Nyabihu", "Ngororero", "Rusizi", "Nyamasheke", 
                 "Rulindo", "Gakenke", "Musanze", "Burera", "Gicumbi", "Rwamagana", "Nyagatere", "Gatsibo", "Kayonza", 
                 "Kirehe", "Ngoma", "Bugesera", "Total"],
    "S A Production (Ton)": [374, 4851, 1570, 7875, 16057, 4613, 4305, 2909, 1898, 2064, 6555, 5732, 9852, 5394, 4902, 
                             6111, 7946, 3627, 9355, 15903, 9378, 9467, 11005, 15889, 50793, 39852, 20181, 21438, 
                             23020, 10561, 333477],
    "S B Production (Ton)": [275, 1797, 504, 2495, 1922, 453, 644, 966, 962, 458, 1634, 468, 2599, 1354, 1706, 892, 
                             640, 315, 440, 834, 1501, 2515, 2909, 2166, 22787, 12249, 5546, 4852, 1947, 13703, 91533],
    "Total Production (Ton)": [649, 6648, 2074, 10370, 17979, 5066, 4949, 3875, 2860, 2522, 8189, 6200, 12451, 6748, 
                               6608, 7003, 8586, 3942, 9795, 16737, 10879, 11982, 13914, 18055, 73580, 52101, 25727, 
                               26290, 24967, 24264, 425010]
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Exclude the "Total" row for district-specific analysis
df_total = df[df['District'] != 'Total']
df_total = df[df['District'] != 'Total']

# Bar plot to visualize maize production by district
plt.figure(figsize=(15, 8))
plt.bar(df_total['District'], df_total['Total Production (Ton)'], color='blue', edgecolor='black')
plt.title('Maize Production by District')
plt.xlabel('District')
plt.ylabel('Total Production (Ton)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Visualize distribution of maize production
plt.figure(figsize=(10, 6))
plt.hist(df['Total Production (Ton)'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Maize Production')
plt.xlabel('Production Value (Ton)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['Total Production (Ton)'])
plt.title('Boxplot of Maize Production')
plt.ylabel('Production Value (Ton)')
plt.show()
```

```python
# Merge maize production data with geospatial data
merged = gdf.set_index('District').join(df.set_index('District'))

# Plot the map
plt.figure(figsize=(12, 8))
merged.plot(column='Total Production (Ton)', cmap='OrRd', legend=True)
plt.title('Maize Production by District in Rwanda (2018)')
plt.show()
```

```python
# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
merged.plot(column='Total Production (Ton)', cmap='OrRd', legend=True, ax=ax)

# Add district names
for x, y, label in zip(merged.geometry.centroid.x, merged.geometry.centroid.y, merged.index):
    ax.annotate(text=label, xy=(x, y), xytext=(3, 3),
                textcoords="offset points", fontsize=8, color='black')

ax.set_title('Maize Production by District in Rwanda (2018)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Check for matching district names
print("Shapefile Districts:", gdf['District'].unique())
print("Data Districts:", df['District'].unique())

# Merge maize production data with geospatial data
merged = gdf.set_index('District').join(df.set_index('District'))

# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
merged.plot(column='Total Production (Ton)', cmap='OrRd', legend=True, ax=ax,
            legend_kwds={'label': "Total Production (Ton)",
                         'orientation': "horizontal"})

# Add district names
for x, y, label in zip(merged.geometry.centroid.x, merged.geometry.centroid.y, merged.index):
    ax.annotate(text=label, xy=(x, y), xytext=(3, 3),
                textcoords="offset points", fontsize=8, color='black')

# Manually extend plot limits to ensure all districts are included
x_buffer = 0.1
y_buffer = 0.1
ax.set_xlim(merged.total_bounds[0] - x_buffer, merged.total_bounds[2] + x_buffer)
ax.set_ylim(merged.total_bounds[1] - y_buffer, merged.total_bounds[3] + y_buffer)

# Customize plot
ax.set_title('Maize Production by District in Rwanda (2018)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```
